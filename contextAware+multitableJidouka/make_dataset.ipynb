```python
import re
def modify_query(query):
    modified_query = re.sub(r"= '([^']*)'", r"LIKE LOWER('%\1%')", query)
    return modified_query
```

```python
data2 = []
for i in range(len(data)):
    lst = []
    lst.append(data[i][0])
    lst.append(modify_query(data[i][1]))
    lst.append(data[i][2])
    lst.append(data[i][3])
    lst.append(modify_query(data[i][4]))
    data2.append(lst)
```
